```
grid = np.empty([height, width, 3], dtype=np.uint8)  # Create empty gird

all_values1 = [] # Will be filled with all values that are not 99999
all_values2 = [] # Will be filled with all values that are not 99999
difference = []

for r, row in enumerate(data_listsDec):
    for c, value in enumerate(row):
      value1 = float(data_listsDec[r][c])
      value2 = float(data_listsJan[r][c])
      if value1 < 99999:
        all_values1.append(value1)
      if value2 < 99999:
        all_values2.append(value2)
      
      if value1 == 99999 and value2 < 1000: #turned to sea
        grid[r, c] = [0, 120, 6]
      elif value1 < 1000 and value2 == 99999: #turned to land
        grid[r, c] = [89, 43, 0]
      elif value1 == 99999 and value2 == 99999:
        grid[r, c] = [0, 0, 0]
      else:
        diff = value2-value1
        difference.append(diff)
        if diff <= -4:
          grid[r, c] = [0, 0, 255] 
        elif diff <= -3:
          grid[r, c] = [0, 107, 255]
        elif diff <= -2:
          grid[r, c] = [0, 142, 255]
        elif diff <= -1:
          grid[r, c] = [0, 186, 255]
        elif diff < 0:
          grid[r, c] = [178, 211, 255]
        elif diff == 0:
          grid[r, c] = [255, 255, 255] 
        elif diff <= 1:
          grid[r, c] = [255, 200, 200]
        elif diff <= 2:
          grid[r, c] = [255, 186, 164] 
        elif diff <= 3:
          grid[r, c] = [255, 172, 147]
        elif diff <= 4:
          grid[r, c] = [255, 143, 112] 
        elif diff <= 5:
          grid[r, c] = [255, 112, 79] 
        elif diff <= 6:
          grid[r, c] = [255, 75, 45] 
        else:
          grid[r, c] = [255, 0, 0] 
```
