#Importing

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

#Finding the Body Mass

Functions

```
def avg_bird_dino(colName, birdOrDino):
  return birdDinoData[colName].loc[birdDinoData["Bird vs Dino"]==birdOrDino].mean() # Returns the average of that column
```

```
def brain_body_ratio(brain, body):
  return brain/body # returns the brain to body ratio
```

Coding

```
femurLen = 315 # mm
bodyMass = 10**(3.1854*np.log10(femurLen)-3.184)/1000 # Gets the body mass in kilograms
brainVol = 56.31 # cm3
print(bodyMass)
```

```
birdDinoData = pd.read_csv("/content/drive/My Drive/Bridge-Up STEM/Unit 5 - Paleo Neuroscience/bird_dino_data.csv") # Opens and reads the csv file
birdBrainVolAvg = avg_bird_dino("Brain Volume (cm3)", "B") # gets the average brain volume for birds
dinoBrainVolAvg = avg_bird_dino("Brain Volume (cm3)", "D") # gets the average brain volume for dinos
```

```
birdBodyMassAvg = avg_bird_dino("Body Mass (kg)", "B") # gets the average body mass for birds
dinoBodyMassAvg = avg_bird_dino("Body Mass (kg)", "D") # gets the average body mass for dinos
```

```
birdBrainBodyAvg = brain_body_ratio(birdBrainVolAvg, birdBodyMassAvg) # gets the average brain to body mass ratio for birds
dinoBrainBodyAvg = brain_body_ratio(dinoBrainVolAvg, dinoBodyMassAvg) # gets the average brain to body mass ratio for dinos
```

```
ourBrainBody = brain_body_ratio(brainVol, bodyMass) # gets the brain to body mass ratio for our unknown animal
```

```
# Prints the values
print("Average Bird Brain to Body Mass Ratio:", birdBrainBodyAvg)
print("Average Dino Brain to Body Mass Ratio:", dinoBrainBodyAvg)
print("Our Brain to Body Mass Ratio:", ourBrainBody)
```

#Finding the Cerebrum to Brain Volume Ratio


Functions

```
def brain_region_ratio(region, brain):
  return region/brain # returns the brain region to brain ratio
```

Coding

```
birdCerebrumAvg = avg_bird_dino("Cerebrum (cm3)", "B") # gets the average cerebrum volume for birds
dinoCerebrumAvg = avg_bird_dino("Cerebrum (cm3)", "D") # gets the average cerebrum volume for dinos
```

```
birdCerBrainAvg = brain_region_ratio(birdCerebrumAvg, birdBrainVolAvg) # gets the average cerebrum to brain ratio for birds
dinoCerBrainAvg = brain_region_ratio(dinoCerebrumAvg, dinoBrainVolAvg) # gets the average cerebrum to brain ratio for dinos
```

```
ourCerBrain = 36.85/56.31 # calculates the cerebrum to brain ratio for our animal
```

```
# Prints the values
print("Average Bird Cerebrum to Brain Ratio:", birdCerBrainAvg)
print("Average Dino Cerebrum to Brain Ratio:", dinoCerBrainAvg)
print("Our Cerebrum to Brain Ratio:", ourCerBrain)
```

#Predictions

```
def compare(value1, value2, value3):
  if abs(value1-value2)<abs(value3-value2):
    return 1
  elif abs(value1-value2)>abs(value3-value2):
    return 2
  else:
    return 0
```

```
closestStrings = ["Equally Close", "Closest to an avian species", "Closest to a non-avian dinosaur"]
closestBrainBodyMass = compare(birdBrainBodyAvg, ourBrainBody, dinoBrainBodyAvg)
print(closestStrings[closestBrainBodyMass]+" based on Brain to Body Mass Ratio")
```

```
closestCerBrain = compare(birdCerBrainAvg, ourCerBrain, dinoCerBrainAvg)
print(closestStrings[closestCerBrain]+" based on Cerebrum to Brain Ratio")
```
