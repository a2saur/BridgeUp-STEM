#Similarities


```python
simRE = 0 #stores the number of similarities
for counter, elephantNuc in enumerate(elephantDNA): 
  rhinoNuc = rhinoDNA[counter]
  if rhinoNuc == elephantNuc:
    simRE += 1

print(simRE)

simRZ = 0 #stores the number of differences
for counter, rhinoNuc in enumerate(rhinoDNA): 
  zebraNuc = zebraDNA[counter]
  if zebraNuc == rhinoNuc:
    simRZ += 1

print(simRZ)

simZE = 0 #stores the number of differences
for counter, elephantNuc in enumerate(elephantDNA): 
  zebraNuc = zebraDNA[counter]
  if zebraNuc == elephantNuc:
    simZE += 1

print(simZE)

perRE = int((simRE/len(rhinoDNA))*100)
perRZ = int((simRZ/len(rhinoDNA))*100)
perZE = int((simZE/len(rhinoDNA))*100)

print("Rhino and Elephant:", perRE, "%")
print("Rhino and Zebra:", perRZ, "%")
print("Elephant and Zebra:", perZE, "%")
```
